```python
import pandas as pd

def inspecteur_gadget(df):
    # Header rapide
    print(f"--- ANALYSE GADGET : {df.shape[0]} lignes | {df.shape[1]} colonnes ---")
    print("-" * 50)
    
    # Construction d'un tableau récapitulatif
    stats = pd.DataFrame({
        'Type': df.dtypes,
        'Manquants': df.isna().sum(),
        '% Manquants': (df.isna().mean() * 100).round(2).astype(str) + '%',
        'Uniques': df.nunique()
    })
    
    print(stats)
    print("-" * 50)
```
